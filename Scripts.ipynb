#### Import packages

```python
import pandas as pd
```

#### Set file paths for input file "Lista Equivalentes" and output files "TCA" and "Equivalentes"

```python
input_path = "Lista Equivalentes.xlsx"
output_path_tca = "TCA.json"
output_path_equivalentes = "Equivalentes.json"
```

#### Import relevant sheets from workbook

TCA

```python
df1 = pd.read_excel(input_path,engine='openpyxl',sheet_name='TCA',usecols='A:AV')
df1.head()
```

```python
# df1.dtypes
```

Equivalentes

```python
df2 = pd.read_excel(input_path,engine='openpyxl',sheet_name='Equivalentes',usecols='A:E')
df2.head()
```

```python
# df2.dtypes
```

```python
df3 = pd.merge(df2,df1,how='inner',left_on='Alimento',right_on='Nome do alimento').groupby(by=['Grupo','Nome'])[['Porcao','Lípidos (g)','Hidratos de carbono (g)','Proteínas (g)','Álcool (g)']].mean()
df3['Energia (kcal)'] = df3['Lípidos (g)'] * 9 + df3['Hidratos de carbono (g)'] * 4 + df3['Proteínas (g)'] * 4 + df3['Álcool (g)'] * 7
df3['Hidratos (g) Porcao'] = df3['Hidratos de carbono (g)'] * df3['Porcao'] / 100
df3['Lípidos (g) Porcao'] = df3['Lípidos (g)'] * df3['Porcao'] / 100
df3['Proteínas (g) Porcao'] = df3['Proteínas (g)'] * df3['Porcao'] / 100
df3['Álcool (g) Porcao'] = df3['Álcool (g)'] * df3['Porcao'] / 100
df3.reset_index(inplace=True)
df3
```

```python
df1.to_json(output_path_tca,orient='records')
df3.to_json(output_path_equivalentes,orient='records')
```
